```python
from stackapi import StackAPI
site = StackAPI('stackoverflow')
site.max_pages = 20
user_list = site.fetch('users')
#print(user_list)
items = user_list['items']
#print("\nInformation of all users as returned from site")
#print(items)
print("\nDetails of users in a tabular form")
print("\nUser ID\t\tAccount ID\tDisplay Name")
for i in items:
    print(str(i['user_id'])+"\t\t"+str(i['account_id'])+"\t\t"+i['display_name'])
```
